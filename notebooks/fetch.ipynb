```python
api = os.getenv("TMDB_API_KEY")
movies = []

for page in range(1, 3):
    url = f"https://api.themoviedb.org/3/movie/popular?api_key={api}&page={page}"
    res = requests.get(url)
    data = res.json()

    for m in data["results"]:
        movies.append({
            "id":               m.get("id"),
            "title":            m.get("title"),
            "overview":         m.get("overview"),
            "genres":           str(m.get("genre_ids")),
            "vote_avg":         m.get("vote_average"),
            "release_data":     m.get("release_date"),
            "popularity":       m.get("popularity"),
            "budget":           0,      # not in this endpoint
        })

movies
```

```python
def fetch_movie_detail(movie_id):
    api = os.getenv("TMDB_API_KEY")
    url = f"https://api.themoviedb.org/3/movie/{movie_id}?api_key={api}"
    res = requests.get(url)
    data = res.json()
    return {
        "budget": data.get("budget", 0),
        "revenue": data.get("revenue", 0)
    }

fetch_movie_detail(1265609)
```
